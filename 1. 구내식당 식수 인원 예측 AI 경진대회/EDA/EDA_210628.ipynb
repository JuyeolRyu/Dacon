## 1. 데이터 불러오기 

```python
df = pd.read_csv("./train.csv")
print(df.shape)
df.head()
```

```python
df.describe()
```

## 2. 시간 정보 정리 

```python
import datetime

df_tmp = pd.DataFrame(df['일자'].copy())
df_tmp['datetime'] = [datetime.datetime.strptime(timestamp, "%Y-%m-%d") for timestamp in df_tmp['일자']]
df_tmp["date"] = df_tmp["datetime"].dt.date
df_tmp["year"] = df_tmp["datetime"].dt.year
df_tmp['quarter'] = df_tmp['datetime'].dt.quarter
df_tmp['weeks'] = df_tmp['datetime'].dt.week
df_tmp["month"] = df_tmp["datetime"].dt.month
df_tmp["weekday"] = df_tmp["datetime"].dt.weekday
df_tmp["day"] = df_tmp["datetime"].dt.day

df_tmp.head()
```

```python
# 원본 df와 병합 
df_target = pd.merge(df_tmp, df, left_on="일자", right_on="일자", how='left')
df_target.drop(['일자'], axis=1, inplace=True)
df_target = df_target.loc[::-1].reset_index(drop=True)

df_target.head(2)
```

## 7. 중식/석식계 구간 나누기 

### (1) 중식계 구간 나누기

<H4> ⚠⚠ 중식계/실근무인원 으로 계산 필요

```python
a = df_target['중식계']
new = pd.cut(a, 4, precision=2)
new.value_counts()
```

```python
group_names = ['중식_아주_낮음', '중식_보통_높음', '중식_아주_높음', '중식_보통_낮음']
```

```python
new = pd.cut(a, 4, precision=2, labels=group_names)
new.value_counts()
```

```python
df_target['중식_Group'] = pd.cut(a, 4, precision=2, labels=group_names)
df_target['중식_Group']
```

```python
df_target.loc[df_target['중식_Group'] == '중식_아주_낮음', '중식_Group_No'] = 0
df_target.loc[df_target['중식_Group'] == '중식_보통_낮음', '중식_Group_No'] = 1
df_target.loc[df_target['중식_Group'] == '중식_보통_높음', '중식_Group_No'] = 2
df_target.loc[df_target['중식_Group'] == '중식_아주_높음', '중식_Group_No'] = 3
df_target.head(2)
```

## 8. 중식/석식 메뉴 분류하기

```python
df_target.columns
```

```python
df_target['중식메뉴'].head(2)
```

```python
df_target['중식메뉴'].str.split(' ')
```

```python
df_target['중식메뉴'].str.split(' ').get(0)
```

```python
len(df_target)
```

```python
menu_list = []

for i in range(len(df_target)):
    tmp = df_target['중식메뉴'].str.split(' ').get(i)
    for j in range(len(tmp)):
        if tmp[j].startswith('(New)'):
            menu_list.append(tmp[j])
        elif tmp[j].startswith('(') or tmp[j] == '':
            continue 
        else:
            if tmp[j] not in menu_list:
                menu_list.append(tmp[j])

print(menu_list)
```

```python
len(menu_list) # 메뉴 개수
```

## 9. 메뉴 리스트 One-hot-encoding (특성공학)

```python
menu_encoding = pd.get_dummies(menu_list)
menu_encoding.head()
```

```python
# concat함수를 이용해 원래 테이블과 합치기 
df_target = pd.concat([df_target, menu_encoding], axis=1)
```
